```python
import speech_recognition as sr
from gtts import gTTS
from IPython.display import Audio, display, Javascript
from google.colab import files
import time
from pydub import AudioSegment

def hablar(texto):
    tts = gTTS(texto, lang="es")
    tts.save("voz.mp3")
    display(Audio("voz.mp3", autoplay=True))

def subir_audio():
    uploaded = files.upload()
    filename = list(uploaded.keys())[0]
    if filename.endswith(".mp3"):
        sound = AudioSegment.from_mp3(filename)
        filename_wav = "audio.wav"
        sound.export(filename_wav, format="wav")
        return filename_wav
    return filename

def reconocer_audio(archivo):
    r = sr.Recognizer()
    with sr.AudioFile(archivo) as source:
        audio = r.record(source)
    try:
        return r.recognize_google(audio, language="es-ES")
    except:
        return ""

def abrir_youtube(busqueda):
    url = f"https://www.youtube.com/results?search_query={busqueda}"
    display(Javascript(f"window.open('{url}', '_blank');"))

def asistente():
    hablar("Sube tu grabación para buscar en YouTube.")
    archivo = subir_audio()
    time.sleep(1)
    hablar("Procesando...")
    texto = reconocer_audio(archivo)
    if texto:
        hablar(f"Buscando {texto} en YouTube.")
        abrir_youtube(texto.replace(" ", "+"))

asistente()
```
